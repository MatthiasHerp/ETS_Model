```python
for i in range(8,31):
    series = pd.Series(data.iloc[:, i])
    series.index = data["date"]

    #making the index a datetime variable to identify the month later in model where monthly dummies are included
    series.index = pd.to_datetime(series.index)

    #defining the training and evaluation set
    y = series[:-365]
    y_predict = series[-365:]
       
    #calculate initial parameters
    Parameters = Initial_Parameter_calculater(y,exog_to_train)
    
    print("Starting Optimizer")
    
    #optimize Model with initial parameters
    Optimal_Parameters = model_optimization(Parameters, y, exog_to_train, before, after, model)
    
    print("got results")
    print(Optimal_Parameters)
    
    print("storing results")
    dict_optimal_parameters[str(i)].append(Optimal_Parameters)
    print("showing calculated results")
    print(dict_optimal_parameters)
```
